```python
import numpy as np
import cv2
import pywt
import math
from wand.image import Image
from skimage.metrics import structural_similarity as SSIM
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm
```

```python
def DWT(value, level):
    decomposition = dict()
    for i in range(1, level+1):
        decomposition[f'LL{i}'], (decomposition[f'LH{i}'], decomposition[f'HL{i}'], decomposition[f'HH{i}']) = pywt.dwt2(value, 'haar')
        value = decomposition[f'LL{i}']
    return decomposition

def iDWT(decomposition, last_LL, level, use_approximation=False):
    value = last_LL
    while level > 0:
        if use_approximation:
            value = pywt.idwt2((value, (np.zeros_like(value), np.zeros_like(value), np.zeros_like(value))), 'haar')
        else:
            value = pywt.idwt2((value, (decomposition[f'LH{level}'], decomposition[f'HL{level}'], decomposition[f'HH{level}'])), 'haar')
        level -= 1
    return value

def SVD(value):
    U, S, Vt = np.linalg.svd(value)
    return U, S, Vt

def iSVD(U, S, Vt):
    value = U.dot(np.diag(S)).dot(Vt)
    return value

def DWT_SVD(value, level):
    decomposition = DWT(value, level)
    decomposition['U'], decomposition['S'], decomposition['Vt'] = SVD(decomposition[f'LL{level}'])
    return decomposition

def embed_watermark(original_decomposition, watermark_decomposition, alpha, use_svd, level):
    if use_svd:
        S_embedded = original_decomposition['S'] + alpha * watermark_decomposition['S']
        LL_embedded = iSVD(original_decomposition['U'], S_embedded, original_decomposition['Vt'])
        return LL_embedded
    else:
        LL_embedded = original_decomposition[f'LL{level}'] + alpha * watermark_decomposition[f'LL{level}']
        return LL_embedded
    
def extract_watermark(original_decomposition, watermark_decomposition, embedded_decomposition, alpha, use_svd, level):
    if use_svd:
        S_extracted = (embedded_decomposition['S'] - original_decomposition['S'])/alpha
        LL_extracted = iSVD(watermark_decomposition['U'], S_extracted, watermark_decomposition['Vt'])
        return LL_extracted
    else:
        LL_extracted = (embedded_decomposition[f'LL{level}'] - original_decomposition[f'LL{level}'])/alpha
        return LL_extracted

def MSE(original, testing):
    return np.mean((original - testing) ** 2)

def evaluate(original, testing, image_type):
    original = original.astype('int32')
    testing = testing.astype('int32')
    print(f'MSE: {MSE(original, testing)}')
    print(f'PSNR: {cv2.PSNR(original, testing)}')
    print(f"SSIM: {SSIM(original, testing, channel_axis = 2 if image_type=='color' else None)}")
    
def readImage(image_path, image_type="grayscale", size=None):
    type = cv2.IMREAD_COLOR if image_type=="color" else cv2.IMREAD_GRAYSCALE
    image = cv2.imread(image_path, type)
    if size == None:
        if(image.shape[1]>image.shape[0]): return cv2.resize(image, (image.shape[0], image.shape[0])).astype('int32')
        elif(image.shape[1]<image.shape[0]): return cv2.resize(image, (image.shape[1], image.shape[1])).astype('int32')
        else: return image.astype('int32')
    else:
        return cv2.resize(image, size[::-1]).astype('int32')
    
def showImages(images):
    num_images = len(images)
    plt.figure(figsize=(20,20))
    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        plt.imshow(images[i], cmap='gray', norm = NoNorm())
    plt.show()
    
def add_noise(image_path, noise_type, attenuate=1):
    if noise_type in ("gaussian", "impulse", "laplacian", "multiplicative_gaussian", "poisson", "random", "uniform"):
        image = Image(filename = image_path)
        image.noise(noise_type, attenuate)
        image.save(filename = image_path)
```

```python
def embed_grayscale_image(original_image, watermark_image, alpha, use_svd=True, level=1):
    # Perform DWT then apply SVD on the last LL sub-band of original image
    original_decomposition = DWT_SVD(original_image, level)

    # Perform DWT then apply SVD on the last LL sub-band of watermark image
    watermark_decomposition = DWT_SVD(watermark_image, level)
    
    # Embed watermark
    LL_embedded = embed_watermark(original_decomposition, watermark_decomposition, alpha, use_svd, level)
    
    # Rebuild embedded image using LL_embedded
    embedded_image = iDWT(original_decomposition, LL_embedded, level)
    
    return embedded_image.astype('int32')
    
def embed_color_image(original_image, watermark_image, alpha, use_svd=True, level=1):
    (B_original, G_original, R_original) = cv2.split(original_image)
    (B_watermark, G_watermark, R_watermark) = cv2.split(watermark_image)
    
    B_embedded = embed_grayscale_image(B_original, B_watermark, alpha, use_svd, level)
    G_embedded = embed_grayscale_image(G_original, G_watermark, alpha, use_svd, level)
    R_embedded = embed_grayscale_image(R_original, R_watermark, alpha, use_svd, level)
    
    embedded_image = cv2.merge([B_embedded, G_embedded, R_embedded])
    return embedded_image.astype('int32')
    
def embed_image(original_image, watermark_image, alpha, use_svd=True, level=1):
    # Grayscale image
    if len(original_image.shape)==2 and len(watermark_image.shape)==2:
        return embed_grayscale_image(original_image, watermark_image, alpha, use_svd, level)
    # Color image
    if len(original_image.shape)==3 and len(watermark_image.shape)==3:
        return embed_color_image(original_image, watermark_image, alpha, use_svd, level)
```

```python
def extract_grayscale_image(original_image, watermark_image, embedded_image, alpha, use_svd=True, level=1, use_approximation=False):
    # Perform DWT then apply SVD on the last LL sub-band of original image
    original_decomposition = DWT_SVD(original_image, level)

    # Perform DWT then apply SVD on the last LL sub-band of watermark image
    watermark_decomposition = DWT_SVD(watermark_image, level)
    
    # Perform DWT then apply SVD on the last LL sub-band of embedded image
    embedded_decomposition = DWT_SVD(embedded_image, level)
    
    # Extract watermark
    LL_extracted = extract_watermark(original_decomposition, watermark_decomposition, embedded_decomposition, alpha, use_svd, level)
    
    # Rebuild extracted watermark image using LL_extracted
    extracted_image = iDWT(watermark_decomposition, LL_extracted, level, use_approximation)
    
    return extracted_image.astype('int32')
    
def extract_color_image(original_image, watermark_image, embedded_image, alpha, use_svd=True, level=1, use_approximation=False):
    (B_original, G_original, R_original) = cv2.split(original_image)
    (B_watermark, G_watermark, R_watermark) = cv2.split(watermark_image)
    (B_embedded, G_embedded, R_embedded) = cv2.split(embedded_image)
    
    B_extracted = extract_grayscale_image(B_original, B_watermark, B_embedded, alpha, use_svd, level, use_approximation)
    G_extracted = extract_grayscale_image(G_original, G_watermark, G_embedded, alpha, use_svd, level, use_approximation)
    R_extracted = extract_grayscale_image(R_original, R_watermark, R_embedded, alpha, use_svd, level, use_approximation)
    
    extracted_image = cv2.merge([B_extracted, G_extracted, R_extracted])
    return extracted_image.astype('int32')
    
def extract_image(original_image, watermark_image, embedded_image, alpha, use_svd=True, level=1, use_approximation=False):
     # Grayscale image
    if len(embedded_image.shape)==2 and len(original_image.shape)==2:
        return extract_grayscale_image(original_image, watermark_image, embedded_image, alpha, use_svd, level, use_approximation)
    # Color image
    if len(original_image.shape)==3 and len(watermark_image.shape)==3:
        return extract_color_image(original_image, watermark_image, embedded_image, alpha, use_svd, level, use_approximation)
```

```python
IMAGE_PATH = "camerawoman.jpg"
WATERMARK_PATH = "watermark.jpg"
EMBEDDED_PATH = 'embedded.jpg'
EXTRACTED_PATH = 'extracted.jpg'
ALPHA = 0.1 # embedding strength
IMAGE_TYPE = "color"
USE_SVD = True
USE_APPROXIMATION = False
LEVEL = 1
ADD_NOISE = True

original_image = readImage(IMAGE_PATH, IMAGE_TYPE)
watermark_image = readImage(WATERMARK_PATH, IMAGE_TYPE, original_image.shape[:2])
showImages((original_image, watermark_image))

embedded_image = embed_image(original_image, watermark_image, ALPHA, USE_SVD, LEVEL)
cv2.imwrite(EMBEDDED_PATH, embedded_image)
if ADD_NOISE:
    add_noise(EMBEDDED_PATH, "impulse", attenuate = 0.01)
    embedded_image = readImage(EMBEDDED_PATH, IMAGE_TYPE)

extracted_image = extract_image(original_image, watermark_image, embedded_image, ALPHA, USE_SVD, LEVEL, USE_APPROXIMATION)
cv2.imwrite(EXTRACTED_PATH, extracted_image)
showImages((embedded_image, extracted_image))

print("Evaluating original image and embedded image:")
evaluate(original_image, embedded_image, IMAGE_TYPE)
print("Evaluating watermark image and extracted image:")
evaluate(watermark_image, extracted_image, IMAGE_TYPE)
```
